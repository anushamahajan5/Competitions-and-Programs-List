**we will split the data**

```python
#creating a train folder, to avoid creation of older again and again we apply if statement
if not os.path.exists("./train"):
  os.mkdir("./train")

  for dir in os.listdir(ROOT_DIR):                  #this will give yes no file
    os.makedirs("./train/"+dir)   

    for img in np.random.choice(a= os.listdir(os.path.join(ROOT_DIR,dir)),
                                size=(math.floor(70/100*number_of_images[dir])-2),
                                replace=False):
       O=os.path.join(ROOT_DIR,dir,img)
       D=os.path.join("./train",dir)
       os.remove(O)
else:
  print("folder exists")                             
   
```

```python
def datafolder(path,split):
  if not os.path.exists("./"+path):
    os.mkdir("./"+path)

    for dir in os.listdir(ROOT_DIR):                            #this will give yes no file
      os.makedirs("./"+path+"/"+dir)   
      for img in np.random.choice(a= os.listdir(os.path.join(ROOT_DIR,dir)),
                                size=(math.floor(split*number_of_images[dir])-2),
                                replace=False):
       O=os.path.join(ROOT_DIR,dir,img)
       D=os.path.join("./"+path,dir)
       shutil.copy(O,D)
       os.remove(O)
  else:
    print("folder exists")                             
   
```

**MODEL BUILD**

```python
from keras.layers import Conv2D, MaxPool2D, Dropout, Flatten, Dense, BatchNormalization, GlobalAveragePooling2D      #keras is the underlying framework for tensorflow
from keras.models import Sequential
from keras.preprocessing.image import ImageDataGenerator   # img_to_array  :to convert image to pixels
import keras
```

```python
#CNN model
model= Sequential()                               #define model as variable

#to add layers to the model we can either define array or use .add method
model.add(Conv2D(filters=16, kernel_size=(3,3),activation='relu',input_shape=(224,224,3))) #kernel size is size of filters  ,we can use padding to prevent change in size

model.add(Conv2D(filters=36, kernel_size=(3,3),activation='relu')) #input shape only for input layer
model.add(MaxPool2D(pool_size=(2,2)))                              #to prevent overfitting

model.add(Conv2D(filters=64, kernel_size=(3,3),activation='relu')) #input shape only for input layer
model.add(MaxPool2D(pool_size=(2,2)))

model.add(Conv2D(filters=128, kernel_size=(3,3),activation='relu')) #input shape only for input layer
model.add(MaxPool2D(pool_size=(2,2)))

model.add(Dropout(rate=0.25))                                      #dropout tells rate at which data decreases

model.add(Flatten())
model.add(Dense(units=64, activation='relu'))
model.add(Dropout(rate=0.25))
model.add(Dense(units=1, activation='sigmoid'))                                     #more dropout layers to prevent overfiting

model.summary()
```

why we have kept filters less first and then increased?
the situation here is similar to capturing a picture on camera. At first, we set a wide angle and look at the wider picture so features are not specific then when we set a specific angle we get to explore more features. when we extract features fromn features i.e we use a feature map we get more information and apply more filters

```python
model.compile(optimizer='adam',loss=keras.losses.binary_crossentropy, metrics=['accuracy'])
```

**preparing our data using data generator**


```python
def preprocessingimages(path):
  """
  input:path
  output:pre processed images
  """
  image_data = ImageDataGenerator(zoom_range=0.2, shear_range=0.2, rescale=1/255, horizontal_flip= True)
  image= image_data.flow_from_directory(directory=path, target_size=(224,224), batch_size=32,class_mode='binary')

  return image
```

```python
path="/content/train"
train_data=preprocessingimages(path)
```

```python
def preprocessingimages2(path):
  """
  input:path
  output:pre processed images
  """
  image_data = ImageDataGenerator( rescale=1/255)
  image= image_data.flow_from_directory(directory=path, target_size=(224,224), batch_size=32,class_mode='binary')

  return image
```

```python
path="/content/test"
test_data=preprocessingimages2(path)
```

```python
path="/content/val"
val_data=preprocessingimages2(path)
```

```python
#early stopping and model check point 
from keras.callbacks import ModelCheckpoint, EarlyStopping

#early stopping
es=EarlyStopping(monitor="val_accuracy", min_delta=0.01, patience=3, verbose=1,mode='auto')

#model check point
mc=ModelCheckpoint(monitor="val_accuracy", min_delta=0.01, filepath="./bestmodel.h5",verbose=1, save_best_only=True, mode='auto')
cd=[es,mc]
```

**MODEL TRAINING**

```python
hs=model.fit_generator(generator=train_data, steps_per_epoch=8, epochs=30,
                       verbose=1,validation_data=val_data ,
                       validation_steps=16, callbacks=cd)
```
